```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Reload the data

```python
file_path = "/content/My Portfolio VNS.csv"
df = pd.read_csv(file_path)
```

# Remove commas and convert 'Investments' to float

```python
# Remove commas and convert 'Investments' to float
df['Investments'] = df['Investments'].astype(str).str.replace(',', '', regex=False).astype(float)

# Extract numeric part from 'P/L(Rs and %)', remove commas, and convert to float
df['PL_Rs'] = (
    df['P/L(Rs and %)']
    .str.extract(r'([-+]?[0-9,.]+)')[0]
    .str.replace(',', '', regex=False)
    .astype(float)
)
```

# Extract numeric part from 'P/L(Rs and %)', remove commas, and convert to float

```python
df['PL_Rs'] = (
    df['P/L(Rs and %)']
    .str.extract(r'([-+]?[0-9,.]+)')[0]
    .str.replace(',', '', regex=False)
    .astype(float)
)
```

# Clean numeric columns

# Separate profit and loss data

```python
df_profit = df[df['PL_Rs'] > 0]
df_loss = df[df['PL_Rs'] < 0]
```

# Plot: Profit Distribution Pie Chart

```python
plt.figure(figsize=(8, 6))
plt.pie(df_profit['PL_Rs'], labels=df_profit['Symbol'], autopct='%1.1f%%', startangle=140)
plt.title('Profit Distribution by Stock Symbol')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

# Plot: Loss Distribution Pie Chart (absolute values)

```python
plt.figure(figsize=(8, 6))
plt.pie(abs(df_loss['PL_Rs']), labels=df_loss['Symbol'], autopct='%1.1f%%', startangle=140)
plt.title('Loss Distribution by Stock Symbol')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
